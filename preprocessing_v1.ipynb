Biến đổi giá trị tag

```python
# Thay NaN bằng 'Không có tag'
df['tag'] = df['tag'].fillna('Không có tag')

# (Tùy chọn) Chuẩn hóa văn bản về chữ thường, xóa khoảng trắng dư
df['tag'] = df['tag'].str.strip().str.lower()

# Kiểm tra lại các giá trị sau khi xử lý
print(df['tag'].value_counts())
```

Trích xuất dữ liệu từ cột rating_numbers

```python
# Hàm để trích xuất overall rating, number of ratings và number of students
def extract_rating_info(rating_str):
    overall_rating = None
    num_ratings = None
    num_students = None
    
    # Trích xuất overall rating (lấy số sau "Xếp hạng: ")
    overall_match = re.search(r'Xếp hạng: (\d,\d)/5', rating_str)
    if overall_match:
        overall_rating = float(overall_match.group(1).replace(',', '.'))
    
    # Trích xuất number of ratings (lấy số trong "(X xếp hạng)")
    ratings_match = re.search(r'\((\d+) xếp hạng\)', rating_str)
    if ratings_match:
        num_ratings = int(ratings_match.group(1))
    
    # Trích xuất number of students (lấy số trong "X học viên")
    students_match = re.search(r'(\d+) học viên', rating_str)
    if students_match:
        num_students = int(students_match.group(1))
    
    return overall_rating, num_ratings, num_students

# Áp dụng hàm trích xuất vào cột rating_number
df[['overall_rating', 'number_of_ratings', 'number_of_students']] = df['rating_number'].apply(
    lambda x: pd.Series(extract_rating_info(x))
)

# Xóa cột rating_number gốc nếu không cần nữa
# df = df.drop(columns=['rating_number'])

# Hiển thị kết quả
print(df[['overall_rating', 'number_of_ratings', 'number_of_students']].head())
```

Trích xuất dữ liệu từ cột instructor

```python
# Hàm để trích xuất thông tin từ cột instructor
def extract_instructor_info(instructor_str):
    instructor_name = None
    instructor_rating = None
    num_reviews = None
    num_students = None
    num_courses = None
    instructor_description = None
    
    # Trích xuất tên instructor (từ đầu đến trước "X,Y xếp hạng giảng viên")
    name_match = re.match(r'^(.*?)(?=\d,\d xếp hạng giảng viên)', instructor_str, re.DOTALL)
    if name_match:
        instructor_name = name_match.group(1).strip()
    
    # Trích xuất instructor rating (lấy số thập phân từ "X,Y xếp hạng giảng viên")
    rating_match = re.search(r'(\d,\d) xếp hạng giảng viên', instructor_str)
    if rating_match:
        instructor_rating = float(rating_match.group(1).replace(',', '.'))
    
    # Trích xuất số lượt đánh giá (lấy số từ "X đánh giá")
    reviews_match = re.search(r'(\d+) đánh giá', instructor_str)
    if reviews_match:
        num_reviews = int(reviews_match.group(1))
    
    # Trích xuất số học viên (lấy số từ "X học viên")
    students_match = re.search(r'(\d+) học viên', instructor_str)
    if students_match:
        num_students = int(students_match.group(1))
    
    # Trích xuất số khóa học (lấy số từ "X khóa học")
    courses_match = re.search(r'(\d+) khóa học', instructor_str)
    if courses_match:
        num_courses = int(courses_match.group(1))
    
    # Trích xuất mô tả (từ sau "X khóa học" đến hết)
    description_match = re.search(r'\d+ khóa học(.*)$', instructor_str, re.DOTALL)
    if description_match:
        instructor_description = description_match.group(1).strip()
    
    return instructor_name, instructor_rating, num_reviews, num_students, num_courses, instructor_description

# Áp dụng hàm trích xuất vào cột instructor
df[['instructor_name', 'instructor_rating', 'instructor_num_reviews', 'instructor_num_students', 
    'instructor_num_courses', 'instructor_description']] = df['instructor'].apply(
    lambda x: pd.Series(extract_instructor_info(x))
)

# Xóa cột instructor gốc nếu không cần nữa
# df = df.drop(columns=['instructor'])

# Hiển thị kết quả
print(df[['instructor_name', 'instructor_rating', 'instructor_num_reviews', 
          'instructor_num_students', 'instructor_num_courses', 'instructor_description']].head())
```

Trích xuất dữ liệu từ cột publish date

```python
# Hàm để trích xuất tháng và năm từ cột publish_date
def extract_publish_date(publish_str):
    month = None
    year = None
    
    # Trích xuất tháng và năm từ chuỗi "Lần cập nhật gần đây nhất MM/YYYY"
    match = re.search(r'Lần cập nhật gần đây nhất (\d+)/(\d{4})', publish_str)
    if match:
        month = int(match.group(1))  # Lấy tháng và đảm bảo là số nguyên
        year = int(match.group(2))   # Lấy năm và đảm bảo là số nguyên
    
    return month, year

# Áp dụng hàm trích xuất vào cột publish_date
df[['publish_month', 'publish_year']] = df['publish_date'].apply(
    lambda x: pd.Series(extract_publish_date(x))
)

# Đảm bảo cột publish_month và publish_year là kiểu số nguyên (int) nếu có giá trị
df['publish_month'] = df['publish_month'].astype('Int64')  # Sử dụng Int64 để hỗ trợ giá trị null
df['publish_year'] = df['publish_year'].astype('Int64')    # Sử dụng Int64 để hỗ trợ giá trị null

# Xóa cột publish_date gốc nếu không cần nữa
# df = df.drop(columns=['publish_date'])

# Hiển thị kết quả
print(df[['publish_month', 'publish_year']].head())
```

Trích xuất dữ liệu từ cột time_spend

```python
def extract_time_in_minutes(time_str):
    # Kiểm tra nếu giá trị là NaN hoặc không phải chuỗi
    if pd.isna(time_str) or not isinstance(time_str, str):
        return None
    
    total_minutes = None
    
    # Trích xuất số giờ và số phút từ chuỗi
    hours_match = re.search(r'(\d+)\s*giờ', time_str)
    minutes_match = re.search(r'(\d+)\s*phút', time_str)
    
    hours = 0
    minutes = 0
    
    # Nếu có giờ, lấy số giờ
    if hours_match:
        hours = int(hours_match.group(1))
    
    # Nếu có phút, lấy số phút
    if minutes_match:
        minutes = int(minutes_match.group(1))
    
    # Chuyển đổi thành tổng số phút
    total_minutes = hours * 60 + minutes
    
    return total_minutes

# Áp dụng hàm trích xuất vào cột time_spend
df['total_minutes'] = df['time_spend'].apply(extract_time_in_minutes)

# Đảm bảo cột total_minutes là kiểu số nguyên (Int64 để hỗ trợ giá trị null)
df['total_minutes'] = df['total_minutes'].astype('Int64')

# Xóa cột time_spend gốc nếu không cần nữa
# df = df.drop(columns=['time_spend'])

# Hiển thị kết quả
print(df[['total_minutes']].head())
```
